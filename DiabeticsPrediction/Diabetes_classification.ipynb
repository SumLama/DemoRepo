# Splitting the data and target

```python
X = df.drop(['Outcome'],axis=1)
Y = df['Outcome']
```

```python
X
```

Splittling data into training and testing set

```python
X_train,X_test,Y_train,Y_test =train_test_split( X, Y, test_size=0.2, random_state=22)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

# Model selection

*   SVC
*   Linear regression




```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
model1 = SVC()
model2 = LogisticRegression()
```

# Modeling on training data

```python
model1.fit(X_train,Y_train)
```

```python
model2.fit(X_train,Y_train)
```

# Fit model in test data

```python
pred1 = model1.predict(X_test)
```

```python
pred1
```

```python
pred2 = model2.predict(X_test)
```

```python
pred2
```

# Evaluating models


```python
from sklearn.metrics import accuracy_score
acc1 =accuracy_score(Y_test,pred1)
acc1
```

```python
acc2 =accuracy_score(Y_test,pred2)
acc2
```

* Since model2 i.e LogisticRegression has more accuracy then model1 i.e SVC , we use model2 for classification

#Evalauating accuracy of choosen model based on evaluation metrics

```python
# Precision score
precisionscore =precision_score(Y_test,pred2)
precisionscore
```
